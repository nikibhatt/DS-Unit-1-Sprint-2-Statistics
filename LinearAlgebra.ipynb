###Multiplying a Vector by a Vector (dot product and cross product) 

Two ways to multiply two vectors are:
  - 1. the scaler or Dot Product (the result is a scaler)
  - 2. the vector or cross product (the result is a vector)

###Dot Product 
The Dot Product gives a **number** as an answer (a "scaler", not a vector)

The Dot Product is written using a central dot:

            a · b
this means the Dot Product of **a** and **b**

You can calculate the Dot Product of two vectors by:

      a · b = |a| × |b| × cos(θ)

Where:

|a| is the magnitude (length) of vector a

|b| is the magnitude (length) of vector b

θ is the angle between a and b

We multiply the length of **a** times the length of **b**, then multiply by the cosine of the angle between **a** and **b**

_Example_ Calculate the dot product of vectors **a** and **b**:

a · b = |a| × |b| × cos(θ)

a · b = 10 × 13 × cos(59.5°)

a · b = 10 × 13 × 0.5075...

a · b = 65.98... = 66 (rounded)

###Cross Product
The Cross Product gives a **vector** as an answer, and is sometimes referred to as the **vector product**
when you multiply 

The cross product **a x b** of two vectors is ** another vector** that is at right angles to both:

The magnitude (length) of the cross product equals the area of a parallelogram with vectors **a** and **b** for sides:

the cross product is:

    - zero in length when vectors **a** and **b** point in the same, or opposite direction

    - reaches maximum length when vectors **a** and **b** are at the right angles

and it can point one way or another

\\
You can calculate the cross product by :

a × b = |a| |b| sin(θ) n

|a| is the magnitude (length) of vector a

|b| is the magnitude (length) of vector b

θ is the angle between a and b

n is the unit vector at right angles to both a and b

\\
So the **length** is: the length of **a** times the length of **b** times the sine of the angle between **a** and **b**

then you multiply by the vector **n** to make sure it heads in the right **direction** (at right angles to both **a** and **b**) 

The cross product could point in the completely opposite direction and still be at right angles to the two other vectors, so we have the:

      "Right Hand Rule"

with you right-hand, point your index finger along vector **a** and point your middle finger along vector **b**: the cross prodcut goes in the direction of your thumb

###Determinant of a Matrix 
The determinant of a matrix is a **special number** that can be calculated from a square matrix 

The determinant tells us things about the matrix that are useful in systems of linear equations and helps us to find the inverse of a matrix it is also useful in calculus and more

The **symbol** for a determinant is two vertical lines on either side

|A| means the determinant of the matrix A

###Calculating the determinant 
First of all the matix must be square (same number of rows and columns) then it is simple math.

###For a 2 x 2 Matrix:
A = $\begin{bmatrix}a & b\\c & d\end{bmatrix}$

The determinant is: |A| = ad - bc

|A| = a x d - b x c

###For a 3 x 3 Matrix :

A = $\begin{bmatrix}a & b & c\\d & e & f\\g & h & i \end{bmatrix}$

the determinant is:

|A| = a(ei -fh) - b(di - fg) + c(dh - eg)

![alt text](https://drive.google.com/uc?id=1zu0VxV_XtHybe_De0siN1N3MM4bwdUkJ)

to work out the determinant of a **3 x 3** matrix:

- multiply **a** by the determinant of the 2 x 2 matrix that is **not in a's** row or column

- likewise for **b** and for **c**

- sum them up and remember the minus in front of the **b**

###In summary :

- for a 2 x 2 matrix the determinant is **ad - bc**

- for a 3 x 3 matrix multiply **a** by the **determinant of the 2 x 2 matrix** that is **not** in **a's** row or column, likewise for **b** and **c** but rememeber that **b** has a negative sign!

- The pattern continues in this way for larger matrices: multiply **a** by the **determinant of the matrix** that is **not** in **a's** row or column, contine in this way across the whole row, but remember the + - + - pattern!
